```python
# === FastAI Image Classifier with ConvNeXt (Full Script - Updated Aug & LR) ===

# Step 0: Imports and Setup
import os
from pathlib import Path
import numpy as np
import torch # สำหรับ torch.argmax
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
from PIL import Image, ImageFile
from fastai.vision.all import *
import timm

ImageFile.LOAD_TRUNCATED_IMAGES = True

# Step 1: Define Paths
print("--- Step 1: Defining Paths ---")
base_dir = Path("/content/drive/MyDrive/NewForData") # << ตรวจสอบ Path นี้ให้ถูกต้อง
train_dir_name = "train"
valid_dir_name = "valid"
test_dir_name = "test"

train_path = base_dir/train_dir_name
valid_path = base_dir/valid_dir_name
test_path = base_dir/test_dir_name

print(f"Base directory: {base_dir}")
print(f"Train directory: {train_path}")
print(f"Valid directory: {valid_path}")
print(f"Test directory: {test_path}")

# (Optional) Step 2: Rename files (if needed)
def rename_files(directory):
    if not directory.is_dir():
        print(f"ไม่พบไดเรกทอรี: {directory}")
        return
    for class_name in os.listdir(directory):
        class_path = directory/class_name
        if class_path.is_dir():
            for filename_obj in class_path.iterdir():
                filename = filename_obj.name
                if filename.lower().endswith(('.jpg', '.png', '.jpeg')):
                    new_filename = filename.replace('-', '_').replace(' ', '_')
                    if new_filename != filename:
                        try:
                            os.rename(class_path/filename, class_path/new_filename)
                        except OSError as e:
                            print(f"เกิดข้อผิดพลาดในการเปลี่ยนชื่อไฟล์ {filename} ใน {class_path}: {e}")

print("\n--- Step 2: Renaming files (if necessary) ---")
if train_path.exists(): rename_files(train_path)
if valid_path.exists(): rename_files(valid_path)
if test_path.exists(): rename_files(test_path)

# Step 3: Prepare DataLoaders for Training/Validation
print("\n--- Step 3: Preparing DataLoaders (dls) ---")
TARGET_DIMS_FASTAI = (224,224)
BATCH_SIZE = 32

if not train_path.is_dir() or not any((train_path/i).is_dir() for i in os.listdir(train_path)):
    raise ValueError(f"ไม่พบไดเรกทอรีสำหรับ train: {train_path} หรือไม่มี class subdirectories")

item_tfms_presizing = Resize(460) # Presizing

# อัปเดต batch_tfms_presizing ตามที่คุณระบุ
print("  Using updated batch_tfms_presizing for Data Augmentation.")
batch_tfms_presizing = [*aug_transforms(size=TARGET_DIMS_FASTAI[0],
                                        min_scale=0.75,
                                        max_rotate=20.0, # ค่าใหม่
                                        max_lighting=0.3,
                                        max_warp=0.1,    # ค่าใหม่
                                        flip_vert=True,
                                        p_affine=0.8,
                                        p_lighting=0.8
                                       ),
                        Normalize.from_stats(*imagenet_stats)]

poop_datablock = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    splitter=GrandparentSplitter(train_name=train_dir_name,valid_name=valid_dir_name),
    get_y=parent_label,
    item_tfms=item_tfms_presizing,
    batch_tfms=batch_tfms_presizing # ใช้ batch_tfms ที่อัปเดตแล้ว
)

dls = None
try:
    dls = poop_datablock.dataloaders(
        base_dir,
        bs=BATCH_SIZE,
        num_workers=2
    )
    print("สร้าง DataLoaders (dls) สำหรับ train/validation สำเร็จแล้ว")
    print(f"  Class ที่พบ: {dls.vocab}")
    print(f"  จำนวนภาพใน Training set: {len(dls.train.dataset)}")
    print(f"  จำนวนภาพใน Validation set: {len(dls.valid.dataset)}")
    NUM_CLASSES = dls.c
    print(f"  จำนวน Class ทั้งหมด (NUM_CLASSES): {NUM_CLASSES}")
    if NUM_CLASSES == 0:
        raise ValueError("จำนวน class เป็น 0, ไม่สามารถสร้างโมเดลได้")

    print("\n  แสดงตัวอย่าง batch จาก dls.train:")
    dls.show_batch(nrows=2,ncols=3,figsize=(8,6))
    plt.show()

except Exception as e_dls:
    print(f"เกิดข้อผิดพลาดในการสร้าง DataLoaders (dls): {e_dls}")
    raise

# Step 4: Prepare Test DataLoader (test_dl)
# (โค้ดส่วนนี้เหมือนเดิม)
print("\n--- Step 4: Preparing Test DataLoader (test_dl) ---")
test_dl = None
if test_path.is_dir():
    test_files = get_image_files(test_path)
    if test_files and len(test_files) > 0:
        try:
            has_subfolders = any((test_path/i).is_dir() for i in os.listdir(test_path))
        except Exception as e_listdir_test:
            print(f"เกิดข้อผิดพลาดในการตรวจสอบ subdirectories ใน {test_path}: {e_listdir_test}")
            has_subfolders = False

        if has_subfolders:
            print(f"  Test directory ({test_path}) มี class subfolders, กำลังสร้าง test_dl แบบมี label...")
            test_specific_block = DataBlock(
                blocks=(ImageBlock, CategoryBlock), get_items=get_image_files,
                splitter=RandomSplitter(valid_pct=0.0, seed=42), get_y=parent_label,
                item_tfms=Resize(TARGET_DIMS_FASTAI[0]),
                batch_tfms=[Normalize.from_stats(*imagenet_stats)]
            )
            try:
                temp_dls_test = test_specific_block.dataloaders(test_path, bs=BATCH_SIZE, shuffle=False, drop_last=False)
                test_dl = temp_dls_test.train
                if hasattr(test_dl, 'vocab') and test_dl.vocab and len(test_dl.vocab) > 0:
                    print(f"  สร้าง test_dl (มี label) สำเร็จ: {len(test_dl.dataset)} ภาพ, vocab: {test_dl.vocab}")
                else:
                    print("  สร้าง test_dl จาก test_path แล้ว แต่ไม่มี vocab/labels ที่ถูกต้อง.")
                    test_dl = None
            except Exception as e_create_labeled_tdl:
                print(f"  ไม่สามารถสร้าง test_dl แบบมี label จาก {test_path}: {e_create_labeled_tdl}")
                test_dl = None

        if test_dl is None:
            print(f"  Test directory ({test_path}) ไม่มี class subfolders หรือสร้างแบบมี label ไม่สำเร็จ, กำลังสร้าง test_dl แบบไม่มี label...")
            if dls:
                try:
                    test_dl = dls.test_dl(test_files, with_labels=False, shuffle=False)
                    if test_dl and hasattr(test_dl, 'items'):
                        print(f"  สร้าง test_dl (ไม่มี label) จาก dls หลัก: {len(test_dl.items)} ภาพ")
                    else: test_dl = None
                except Exception as e_unlabeled_tdl :
                    print(f"  เกิดข้อผิดพลาดในการสร้าง test_dl แบบไม่มี label: {e_unlabeled_tdl}")
                    test_dl = None
            else:
                print("  ไม่สามารถสร้าง test_dl แบบไม่มี label ได้ เนื่องจาก dls หลักยังไม่ได้ถูกสร้าง")
    else:
        print(f"  ไม่พบไฟล์ภาพในไดเรกทอรี test: {test_path}")
else:
    print(f"  ไม่พบไดเรกทอรี test: {test_path}")

if not test_dl:
    print("  สรุป: ไม่สามารถสร้าง Test DataLoader (test_dl) ได้")


# Step 5: Create Learner with ConvNeXt and potentially Weighted Loss
print("\n--- Step 5: Creating Learner with ConvNeXt ---")
learn = None
if dls and NUM_CLASSES > 0:
    checkpoint_dir_convnext = base_dir/'convnext_models_checkpoint_improved'
    checkpoint_dir_convnext.mkdir(parents=True, exist_ok=True)
    best_model_convnext_fname = 'best_convnext_model_improved'

    callbacks_list = [
        SaveModelCallback(monitor='valid_loss', fname=best_model_convnext_fname, reset_on_fit=True),
        EarlyStoppingCallback(monitor='valid_loss', min_delta=0.001, patience=15),
        ReduceLROnPlateau(monitor='valid_loss', factor=0.2, patience=7)
    ]

    if NUM_CLASSES == 2: f1_metric = F1Score()
    else: f1_metric = F1Score(average='macro')
    metrics_list = [f1_metric, accuracy, error_rate]

    convnext_model_arch = 'convnext_tiny'
    print(f"  กำลังใช้สถาปัตยกรรม: {convnext_model_arch}")

    loss_func_custom = FocalLossFlat()
    print(f"  กำลังใช้ Loss Function: FocalLossFlat")

    try:
        learn = vision_learner(
            dls, convnext_model_arch,
            metrics=metrics_list, cbs=callbacks_list,
            path=checkpoint_dir_convnext, model_dir=".",
            ps=0.5,
            loss_func=loss_func_custom
        )
        print("  สร้าง Learner (ConvNeXt) สำเร็จแล้ว")
    except Exception as e_learner:
        print(f"  เกิดข้อผิดพลาดในการสร้าง Learner: {e_learner}")
        raise
else:
    print("  ไม่สามารถสร้าง Learner ได้ เนื่องจาก dls หรือ NUM_CLASSES ไม่ถูกต้อง")

# Step 6: Find Learning Rate and Train Model
if learn:
    print("\n--- Step 6: Finding Learning Rate and Training ---")
    # กำหนด base_learning_rate โดยตรงตามที่คุณระบุ
    base_learning_rate = 0.000691
    print(f"  ใช้ base_learning_rate ที่กำหนดเอง: {base_learning_rate}")

    # ส่วนของ lr_find() สามารถคอมเมนต์ออกได้ถ้าไม่ต้องการรันทุกครั้ง หรือใช้เพื่อตรวจสอบเท่านั้น
    # try:
    #     print("  กำลังรัน Learning Rate Finder (เพื่อตรวจสอบ)...")
    #     lr_find_res = learn.lr_find(suggest_funcs=(valley, slide))
    #     if lr_find_res.valley: print(f"    LR Finder - Valley: {lr_find_res.valley}")
    #     if lr_find_res.slide: print(f"    LR Finder - Slide: {lr_find_res.slide}")
    # except Exception as e_lr:
    #     print(f"  เกิดข้อผิดพลาดระหว่าง lr_find(): {e_lr}")

    num_fine_tune_epochs = 40
    initial_freeze_epochs = 7

    print(f"  จะเริ่ม fine_tune ด้วย base_lr = {base_learning_rate}, freeze_epochs={initial_freeze_epochs}, num_epochs={num_fine_tune_epochs}")
    try:
        learn.fine_tune(num_fine_tune_epochs, base_lr=base_learning_rate, freeze_epochs=initial_freeze_epochs)
        print("\n  การฝึกโมเดลเสร็จสิ้น")

        print("\n  แสดงผลการฝึก (Loss และ Metrics):")
        learn.recorder.plot_loss(); plt.suptitle("Training & Validation Loss (ConvNeXt - Updated)"); plt.show()
        learn.recorder.plot_metrics(); plt.suptitle("Metrics (ConvNeXt - Updated)"); plt.show()

        print(f"\n  กำลังโหลดโมเดลที่ดีที่สุด ('{best_model_convnext_fname}.pth')...")
        learn.load(best_model_convnext_fname)
        print("  โหลดโมเดลที่ดีที่สุดสำเร็จแล้ว")

    except Exception as e_train:
        print(f"  เกิดข้อผิดพลาดระหว่างการฝึกโมเดล: {e_train}")
else:
    print("  ไม่สามารถเริ่มการฝึกได้ เนื่องจาก Learner (learn) ไม่ได้ถูกสร้าง")


# Step 7: Evaluate on Test Set
# (โค้ดส่วนนี้เหมือนเดิม แต่จะประเมินผลโมเดล ConvNeXt ที่ปรับปรุงแล้ว)
if learn and test_dl:
    print("\n--- Step 7: Evaluating on Test Set ---")
    has_test_labels = hasattr(test_dl, 'vocab') and test_dl.vocab and len(test_dl.vocab) > 0
    if has_test_labels:
        print("  Test DataLoader มี labels, กำลังทำการประเมินผล...")
        try:
            print("\n  ผลการประเมินโดยรวมบน Test Set (learn.validate()):")
            validation_results_test = learn.validate(dl=test_dl)
            metric_names_test = ['loss'] + [m.name for m in learn.metrics]
            for name, value in zip(metric_names_test, validation_results_test):
                val_disp = value.item() if isinstance(value, torch.Tensor) else value
                disp_n = name.replace('_', ' ').capitalize()
                print(f"    {disp_n}: {val_disp*100:.2f}%" if name not in ['loss'] else f"    {disp_n}: {val_disp:.4f}")

            interp = ClassificationInterpretation.from_learner(learn, dl=test_dl)

            print("\n  Confusion Matrix (Test Set):")
            interp.plot_confusion_matrix(figsize=(10,8), cmap="Blues", normalize=False, plot_txt=True)
            plt.xticks(rotation=45, ha='right'); plt.yticks(rotation=0)
            plt.title("Confusion Matrix (Test Set - ConvNeXt Updated)"); plt.tight_layout(); plt.show()

            print("\n  Classification Report (Test Set):")
            y_probs, y_true = learn.get_preds(dl=test_dl)
            y_pred = torch.argmax(y_probs, dim=1).numpy()
            y_true_np = y_true.numpy()
            class_names_rep = learn.dls.vocab

            unique_labels_data = np.unique(np.concatenate((y_true_np, y_pred)))
            num_unique_labels_data = np.max(unique_labels_data) + 1 if len(unique_labels_data) > 0 else 0
            if num_unique_labels_data > 0 and class_names_rep and len(class_names_rep) > 0:
                labels_rep = np.arange(num_unique_labels_data)
                target_names_rep = list(class_names_rep) if len(class_names_rep) >= num_unique_labels_data else None
                if target_names_rep and len(target_names_rep) > num_unique_labels_data:
                     target_names_rep = target_names_rep[:num_unique_labels_data]

                report = classification_report(y_true_np, y_pred, labels=labels_rep, target_names=target_names_rep, zero_division=0)
                print(report)
            elif num_unique_labels_data > 0:
                print("    ไม่สามารถหา class_names, แสดง report โดยไม่มี target names.")
                report_no_n = classification_report(y_true_np, y_pred, labels=np.arange(num_unique_labels_data), zero_division=0)
                print(report_no_n)
            else:
                print("    ไม่พบ labels ในข้อมูล Test Set (y_true, y_pred)")

        except Exception as e_eval:
            print(f"  เกิดข้อผิดพลาดระหว่างการประเมินผลบน Test Set: {e_eval}")
    else:
        print("  Test DataLoader ไม่มี labels หรือยังไม่ได้ถูกสร้าง, จะข้ามการประเมินผล chi tiết")
elif learn and not test_dl:
    print("\n--- Step 7: No Test DataLoader (test_dl) to evaluate ---")
elif not learn:
    print("\n--- Step 7: Learner not available, skipping evaluation ---")


# Step 8: Save Final Model
if learn:
    print("\n--- Step 8: Saving Final Model (ConvNeXt - Updated) ---")
    save_model_final_dir = base_dir/"saved_models_convnext_final_updated"
    save_model_final_dir.mkdir(parents=True, exist_ok=True)
    final_model_convnext_name = "convnext_poop_classifier_updated_v1"

    try:
        best_model_source = checkpoint_dir_convnext/f"{best_model_convnext_fname}.pth"
        if best_model_source.exists():
            destination_path = save_model_final_dir/f"{final_model_convnext_name}.pth"
            shutil.copyfile(best_model_source, destination_path)
            print(f"  โมเดลที่ดีที่สุด ('{best_model_convnext_fname}.pth') ถูกคัดลอกไปยัง: {destination_path}")
        else:
            print(f"  ไม่พบ Best model ({best_model_source}), จะ save โมเดลปัจจุบันแทน...")
            learn.save(final_model_convnext_name, path=save_model_final_dir)
            print(f"  โมเดลปัจจุบันถูกบันทึกที่: {save_model_final_dir/final_model_convnext_name}.pth")
    except Exception as e_save_final:
        print(f"  เกิดข้อผิดพลาดในการบันทึกโมเดลสุดท้าย: {e_save_final}")
else:
    print("\n--- Step 8: Learner not available, skipping model saving ---")

print("\n=== สคริปต์ทั้งหมดเสร็จสิ้น ===")
```
